```python
df.dropna(subset=['embarked'], axis=0, inplace=True)
df.dropna(subset=['embark_town'], axis=0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

```python
df.columns = df.columns.str.upper()
df
```

```python
sns.boxplot(x='CLASS',y='AGE',data=df)
```

We can see that the mean age in each of the class is as follows.. what we can do is replace the null value im the age column with the mean age of that particular class..
- By the the above boxplot we can say that the mean age of First class is approximately 37, Second class is 29 and Third class is 24

```python
def impute_age(cols):
    AGE = cols[0]
    CLASS = cols[1]
    
    if pd.isnull(AGE):

        if CLASS == 1:
            return 37

        elif CLASS == 2:
            return 29

        else:
            return 24

    else:
        return AGE
```

```python
df['AGE'] = df[['AGE','CLASS']].apply(impute_age,axis=1)
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

The DECK column has too many missing values.. 688 out 891 are missing values.. hence it makes sense to drop this particulat feature

```python
df.drop('DECK',axis=1,inplace=True)
```
